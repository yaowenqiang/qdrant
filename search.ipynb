```python
from qdrant_client import QdrantClient
from qdrant_client.http import models
client = QdrantClient('localhost', port=6333)
client.get_collections()
```

```python
from random import uniform

# Simple search


search_vec = [uniform(-0.999, 0.999) for x in range(512)]
client.search(
    collection_name='wolt_food',
    query_vector=(search_vec),
    limit=1, with_vectors=True
)
```

```python
# Search with Filters

search_vec = [-0.027882958, 0.044092737]
client.search(
    collection_name='wolt_food',
    query_vector=(search_vec),
    limit=3,
    query_filter=models.Filter(
        must=[
            models.FieldCondition(
                key='name',
                match=models.MatchValue(
                    value='Ayran',
                )
            ),
            models.FieldCondition(
                key='description',
                match=models.MatchValue(
                    value='',
                )
            )
        ]
    ),
)
```

```python
# Search Filter clause - should
search_vec = []
client.search(
    collection_name='wolt_food',
    query_vector=search_vec,
    limit=3,
    query_filter=models.Filter(
        should=[
            models.FieldCondition(
                key='name',
                match=models.MatchValue(
                    value='Ryran'
                )
            ),
            models.FieldCondition(
                key='description',
                match=models.MatchValue(
                    value='Hello'
                )   
            )
        ]
    )
)
```

```python
# Search Filter clause - mustnot
search_vec = []
client.search(
    collection_name='wolt_food',
    query_vector=search_vec,
    limit=3,
    query_filter=models.Filter(
        must_not=[
            models.FieldCondition(
                key='name',
                match=models.MatchValue(
                    value='Ryran'
                )
            ),
            models.FieldCondition(
                key='description',
                match=models.MatchValue(
                    value='Hello'
                )   
            )
        ]
    )
)
```

```python
# Search Filter clause - mixing multiple condition
search_vec = []
client.search(
    collection_name='wolt_food',
    query_vector=search_vec,
    limit=3,
    query_filter=models.Filter(
        must=[
            models.FieldCondition(
                key='name',
                match=models.MatchValue(
                    value='Ryran'
                )
            )],
        must_not=[
             models.FieldCondition(
                key='description',
                match=models.MatchValue(
                    value='Hello',
                )
            )
        ]
    )
)
```

```python
# Search with filters - filtering conditions - match any
```

```python
# Search Filter clause - mixing multiple condition
search_vec = []
client.search(
    collection_name='wolt_food',
    query_vector=search_vec,
    limit=3,
    query_filter=models.Filter(
        must=[
            models.FieldCondition(
                key='name',
                match=models.MatchAny(
                    any=['Ryran', '7gin Ayran Natur']
                )
            )]
    )
)
```

```python
# Search Filter clause - mixing multiple condition
search_vec = []
client.search(
    collection_name='wolt_food',
    query_vector=search_vec,
    limit=3,
    query_filter=models.Filter(
        must=[
            models.FieldCondition(
                key='name',
                match=models.MatchExcept(
                    **{'except': ['Ayran', ]}
                )
            )]
    )
)
```

```python
# Search Filter clause - nested key
search_vec = []
client.search(
    collection_name='wolt_food',
    query_vector=search_vec,
    limit=3,
    query_filter=models.Filter(
        must=[
            models.FieldCondition(
                key='cafe.address',
                match=models.MatchValue(
                    value='address'
                )
            )]
    )
)
```

```python
# Search Filter clause - match text
search_vec = []
client.search(
    collection_name='wolt_food',
    query_vector=search_vec,
    limit=3,
    query_filter=models.Filter(
        must=[
            models.FieldCondition(
                key='cafe.address',
                match=models.MatchText(
                   text="text
                )
            )]
    )
)
```
